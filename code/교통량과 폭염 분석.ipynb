# 2023.06 비교 (서울)

```python
traffic = pd.read_csv('교통량/06월 서울시 교통량 조사자료(2023).csv', encoding='utf-8')
traffic.head()
```

```python
traffic['일자'] = traffic['일자'].apply(str)
traffic['일자'] = pd.to_datetime(traffic['일자'], format='%Y-%m-%d')
```

```python
heat6 = df[(df['일시'].dt.year==2023) & (df['일시'].dt.month==6) &(df['지점']=='서울(108)') ]
```

```python
heat6.reset_index(inplace=True)
heat6 = heat6.drop('index', axis=1)
```

```python
heat6['교통량'] = np.NaN
```

```python
a = heat6[heat6['폭염특보(O/X)']=='O']
a
```

폭염 특보였던 3일 2023-06-18, 2023-06-19, 2023-06,30

```python
traffic['total'] = traffic['0시']+traffic['1시']+traffic['2시']+traffic['3시']+traffic['4시']+traffic['5시']+traffic['6시']+traffic['7시']+traffic['8시']+traffic['9시']+\
                       traffic['10시']+traffic['11시']+traffic['12시']+traffic['13시']+traffic['14시']+traffic['15시']+traffic['16시']+traffic['17시']+traffic['18시']+traffic['19시']+\
                      traffic['20시']+traffic['21시']+traffic['22시']+traffic['23시']
```

```python
traffic
```

```python
traffic = traffic.dropna()
traffic.isnull().sum()
```

```python
traffic = traffic.drop(['요일', '지점명', '지점번호', '방향', '구분'], axis=1)
```

```python
traffic.sort_values(by=['total'], ascending=False)
```

```python
for i in range(1,31):
    x = traffic[traffic['일자'].dt.day==i]
    heat6['교통량'][i-1] = x.sum()['total']
```

```python
heat6
```

```python
fig, ax1 = plt.subplots(figsize=(40, 10))

sns.barplot(x=heat6['일시'], y=heat6['교통량'])

# 그래프 제목과 축 레이블 설정
plt.xlabel('Date', fontdict={'fontsize':25})
plt.ylabel('Traffic Count', fontdict={'fontsize':25})

#plt.xticks(a.index[::1], a.index[::1], fontsize=20)
plt.yticks(fontsize=20)
plt.title('2023 06월 날짜별 교통량 합계', fontdict = {'fontsize' : 40, 'fontweight': 'bold'})
plt.rcParams['font.size'] = 20

plt.show()
```
